```python
def conv_to_char(arr):
    lis = list(arr)
    k = ""
    for i in lis:
        k += str(chr(i))
    return k
```

```python
files_192 = []
files_224 = []
files_331 = []
files_512 = []

for i in os.listdir(location_192):
    g = location_192 + "/" + i
    files_192.append(g)

for i in os.listdir(location_224):
    g = location_224 + "/" + i
    files_224.append(g)

for i in os.listdir(location_331):
    g = location_331 + "/" + i
    files_331.append(g)

for i in os.listdir(location_512):
    g = location_512 + "/" + i
    files_512.append(g)
```

```python
image_description = {
    "id" : tf.io.FixedLenFeature([],dtype=tf.string),
    "class" : tf.io.FixedLenFeature([],dtype=tf.int64),
    "image" : tf.io.FixedLenFeature([],dtype=tf.string)
}

y = lambda x : tf.io.parse_single_example(x,image_description)
```

```python
def retrieve_images(img_file,y):

    begin = img_file.rindex("-") + 1
    end = img_file.rindex(".tfrec")

    no_of_images = int(img_file[begin:end])

    img_tfrec = tf.data.TFRecordDataset(img_file)

    img = img_tfrec.map(y)

    ctr = 1

    id = []
    label = []
    image = []

    for dr in img.take(no_of_images):
        id.append(conv_to_char(dr["id"].numpy()))
        label.append(dr["class"].numpy())
        image.append(tf.io.decode_image(dr["image"].numpy(),dtype="float64"))
        print(ctr)
        ctr += 1

    return id,label,image
```
