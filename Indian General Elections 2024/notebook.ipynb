```python
no_of_seats_won_by_each_party = pd.DataFrame(df_2["Leading Party"].value_counts()).reset_index()
no_of_seats_won_by_each_party.columns = ["Leading Party","No of Seats Won"]
no_of_seats_won_by_each_party = no_of_seats_won_by_each_party.sort_values(by="No of Seats Won",ascending=False)
no_of_seats_won_by_each_party.head()
```

```python
reg = requests.get(url="https://en.wikipedia.org/wiki/List_of_National_Democratic_Alliance_members")
bs = BeautifulSoup(reg.content,"html.parser")

tab = bs.find("table")

p = []

for i in tab.find_all("tr"):
    q = []
    l = i.text.strip().split("\n")
    for j in l:
        if j!="":
            q.append(j)
    

    p.append(q)

p = p[3:-1]

po = []

for i in p:
    if i == ["State party"]:
        continue
    elif i == ["Unrecognised party"]:
        continue
    elif i == ["Unrecognised Party"]:
        continue
    elif i == ["National party"]:
        continue
    po.append(i)  

nda_members = []

for i in po:
    nda_members.append(i[0])

nda_members = np.array(nda_members)
nda_members = np.sort(nda_members)
```

```python
req = requests.get(url="https://en.wikipedia.org/wiki/List_of_Indian_National_Developmental_Inclusive_Alliance_members")

soup = BeautifulSoup(req.content,"html.parser")

table = soup.find("table")
p = []
for i in table.find_all("tr"):
    q = []
    lk = i.text.strip().split("\n")
    for j in lk:
        if j!='':
            q.append(j)
    p.append(q)

p = p[1:43]
o = 0

india_members = []
for i in p:
    india_members.append(i[1])
    o += 1

india_members = np.array(india_members)
india_members = np.unique(india_members)
```
